#### 폴더에 있는 파일 개수 세기(라벨과 이미지가 저장되었는지 확인)

#### xml 파일 지우기

```python
# Define the base directory for labels
base_dir = '/content/drive/MyDrive/oss_final_project/oss_final_project2/labels'

# Loop through directories from Bbox_0041 to Bbox_0097
for i in range(41, 98):
    dir_path = os.path.join(base_dir, f'Bbox_{i:04d}')

    # Ensure the directory exists
    if not os.path.exists(dir_path):
        print(f'Directory {dir_path} does not exist. Skipping...')
        continue

    # List all XML files in the directory
    xml_files = [file for file in os.listdir(dir_path) if file.endswith('.xml')]

    # Delete each XML file
    for xml_file in xml_files:
        xml_file_path = os.path.join(dir_path, xml_file)
        os.remove(xml_file_path)
        print(f'Deleted {xml_file_path}')
```

#### 정리된 label과 image를 my_data/train으로 옮기기

```python
# Define the base directories
base_src_dir = '/content/drive/MyDrive/oss_final_project/oss_final_project2/images'
base_dst_dir = '/content/drive/MyDrive/oss_final_project/oss_final_project2/my_data/train/images'

# Ensure the destination directory exists
if not os.path.exists(base_dst_dir):
    os.makedirs(base_dst_dir)

#  Loop through directories from Bbox_0041 to Bbox_0099
for i in range(41, 100):
    src_dir = os.path.join(base_src_dir, f'Bbox_{i:04d}')

    # Ensure the source directory exists
    if not os.path.exists(src_dir):
        print(f'Source directory {src_dir} does not exist. Skipping...')
        continue

    # List all files in the source directory
    files = os.listdir(src_dir)

    # Move each file to the destination directory
    for file in files:
        src_file_path = os.path.join(src_dir, file)
        dst_file_path = os.path.join(base_dst_dir, file)

        # Move the file
        shutil.move(src_file_path, dst_file_path)
        print(f'Moved {src_file_path} to {dst_file_path}')
```
